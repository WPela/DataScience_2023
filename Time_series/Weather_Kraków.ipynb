```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = 'C:/GitHub_repos/DataScience/Time_series/Data/combined/combined_2023.csv'
```

```python
df = pd.read_csv(file, encoding = 'cp1250')
```

```python
print(df)
```

```python
df = df[df['Nazwa stacji'].str.contains('KRAKÓW', case=False, na=False)]
```

```python
df.groupby(['Nazwa stacji']).count()
```

```python
df['Datetime'] = pd.to_datetime(df['Rok'].astype(str) + '-' +
                                 df['Miesiac'].astype(str).str.zfill(2) + '-' +
                                 df['Dzien'].astype(str).str.zfill(2) + ' ' +
                                 df['Godzina'].astype(str).str.zfill(2) + ':00')
```

```python
print(df)
```

```python
df.dtypes
```

```python
x = df['Datetime']
y1 = df[df['Nazwa stacji'] == 'KRAKÓW-OBSERWATORIUM']['Temperatura powietrza [°C]']
```

```python
plt.plot(x, y1)
plt.show()
```
